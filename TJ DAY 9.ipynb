-   PAYMENTS : Amount of Payment done by user
-   MINIMUM_PAYMENTS : Minimum amount of payments made by user

```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# rescaling
cc_k = cc_nc[['BALANCE','CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE',
       'CREDIT_LIMIT_RANGE_5k-10k', 'CREDIT_LIMIT_RANGE_10k-15k',
       'CREDIT_LIMIT_RANGE_15k-20k', 'CREDIT_LIMIT_RANGE_20k-30k',
       'TOT_TRANSACTION', 'TOT_TRX', 'TENURE_RANGE_8-10', 'TENURE_RANGE_10-11',
       'TENURE_RANGE_12']]

# instantiate
scaler = StandardScaler()

# fit_transform
cc_k_scaled = scaler.fit_transform(cc_k)
cc_k_scaled.shape
```

```python
cc_k_scaled = pd.DataFrame(cc_k_scaled)
cc_k_scaled.columns = ['BALANCE','CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE',
       'CREDIT_LIMIT_RANGE_5k-10k', 'CREDIT_LIMIT_RANGE_10k-15k',
       'CREDIT_LIMIT_RANGE_15k-20k', 'CREDIT_LIMIT_RANGE_20k-30k',
       'TOT_TRANSACTION', 'TOT_TRX', 'TENURE_RANGE_8-10', 'TENURE_RANGE_10-11',
       'TENURE_RANGE_12']
cc_k_scaled.head()
```

# Modelling

```python
# k-means with some arbitrary k
kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(kmeans.labels_)
```

```python
kmeans.labels_
```

# Finding the Optimal Number of Clusters


# SSD

```python
# elbow-curve/SSD
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(cc_k_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
# ssd
plt.plot(ssd)
```

```python
# silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(cc_k_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(cc_k_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
# final model with k=3
kmeans = KMeans(n_clusters=3, max_iter=50)
kmeans.fit(cc_k_scaled)
```
